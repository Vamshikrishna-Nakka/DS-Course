> **Sampling**
>
> Sampling in statistics refers to the process of selecting a subset of
> individuals or items from a larger population to gather data and make
> inferences about the entire population. Instead of collecting data
> from every individual in the population, sampling allows statisticians
> to study a representative sample, which is a smaller and more
> manageable subset of the population. By studying the sample, one can
> make inferences about the population as a whole.
>
> **Population:**
>
> <img
> src="attachment:vertopal_feff489cada24bbeb2a456d3f51870e6/media/image281.png" />
> The population refers to the entire group of individuals or items that
> the researcher is interested in studying. It can be finite or
> infinite.
>
> **Sample:**
>
> <img
> src="attachment:vertopal_feff489cada24bbeb2a456d3f51870e6/media/image139.png" />
> A sample is a subset of the population that is selected for data
> collection and analysis. The goal is for the sample to be
> representative of the population, so that conclusions drawn from the
> sample can be generalized to the larger population.
>
> **Sampling Frame:**
>
> <img
> src="attachment:vertopal_feff489cada24bbeb2a456d3f51870e6/media/image23.png" />
> A sampling frame is a list or representation of all the individuals or
> items in the population from which the sample will be selected. It
> provides a basis for selecting the sample and should ideally include
> all members of the population.
>
> **Sampling Methods:**
>
> Simple Random Sampling:
>
> <img
> src="attachment:vertopal_feff489cada24bbeb2a456d3f51870e6/media/image23.png" />
> Every member of the population has an equal chance of being selected
> for the sample. This can be done with or without replacement.
>
> Stratified Sampling:
>
> <img
> src="attachment:vertopal_feff489cada24bbeb2a456d3f51870e6/media/image282.png" />
> The population is divided into homogeneous subgroups called strata,
> and a sample is selected from each stratum proportionate to its size
> or importance.
>
> Cluster Sampling:
>
> <img
> src="attachment:vertopal_feff489cada24bbeb2a456d3f51870e6/media/image283.png" />
> The population is divided into clusters or groups, and a subset of
> clusters is randomly selected. Then, all members within the selected
> clusters are included in the sample.
>
> Systematic Sampling:
>
> <img
> src="attachment:vertopal_feff489cada24bbeb2a456d3f51870e6/media/image284.png" />
> The population is ordered, and individuals or items are selected at
> regular intervals, such as every 10th person or every 5th item.
