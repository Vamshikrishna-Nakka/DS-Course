Machine learning (ML) is a branch or subfield of artificial intelligence (AI) deals with the implementation of computational algorithms that improve performance upon experience.

a ML system learns from data. ML approaches include three types of knowledge acquisition: supervised, unsupervised and 
reinforcement learning.

The list of algorithms in each subfield is not exhaustive but instead are the most popular algorithms in each subfield.



### Supervised
In which the algorithm is trained on a labeled dataset, which means that the output(or target)variable is already known. The goal of supervised learning is to learn a function that can accurately predict the output variable based on the input variables.

This is divided into two main categories:

1. Regression: This algorithms are used to predict the continuous values such as price, salary, grade, age, etc.  Based on below table students and there study hours depend on there grade. so grade is independent varibale and student, hours dependent variable.
