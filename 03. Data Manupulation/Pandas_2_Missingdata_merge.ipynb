```python
df.set_index('day',inplace=True)
```

```python
new_df1=df.fillna(0) #fillna(value=None, method=None, axis=None, inplace=False, limit=None, downcast=None)
new_df1              #method=‘backfill’, ‘bfill’, ‘pad’, ‘ffill’, None
```

```python
new_df=df.fillna({
    'temperature':0,
    'windspeed':0,
    'event':'noevent'   # instead of 0 for some columns we can fill required
})
new_df
```

```python
new_df=df.fillna(method='ffill',axis='columns')  #forwardfill,backwardfill
new_df   # on columns the forward fill is happending on rows
```

```python
new_df=df.fillna(method='ffill',limit=1)
new_df #we use ffill it fills forward value that will fill only 1 if in a columns 3 empty but fills only 2 remaining 2 Nan
```

```python
d1=pd.DataFrame({
    'city':['mumbai','chennai','delhi'],
    'temperature':[21,12,32]
})
d2=pd.DataFrame({
    'city':['chennai','Hyderabad'],
    'humidity':[89,23]
})
df3=pd.merge(d1,d2,on='city')   #inner join by default
df3
```

```python
df3=pd.merge(d1,d2,on='city',how='outer',indicator=True) #inner,outer,left,right
df3  #indicator provide an extra column showing the values included
```

```python
d1=pd.DataFrame({
    'city':['mumbai','chennai'],
    'temperature':[21,12],
    'humidity':[60,44]
})

d2=pd.DataFrame({
    'city':['pune','chennai'],
    'temperature':[55,56],
    'humidity':[89,23]
})
df3=pd.merge(d1,d2,on='city',how='outer',suffixes=('_left','_right'))#here it shows_x _y since both df have same column names, if we need to pass we
                                         #use suffixes
df3
```

```python
import pandas as pd
df = pd.DataFrame({
    'brand': ['Yum Yum', 'Yum Yum', 'Indomie', 'Indomie', 'Indomie'],
    'style': ['cup', 'cup', 'cup', 'pack', 'pack'],
    'rating': [4, 4, 3.5, 15, 5]
})
# df.duplicated() #duplicated(subset=None, keep='first') #keep{‘first’, ‘last’, False}, default ‘first’
df
```

```python
df.duplicated() # row same or not
```

```python
df.duplicated(keep='last') #last occurrence of each set of duplicated values is set on False and all others on True.
```

```python
#DataFrame.drop_duplicates(subset=None, keep='first', inplace=False, ignore_index=False)
df = pd.DataFrame({
    'brand': ['Yum Yum', 'Yum Yum', 'Indomie', 'Indomie', 'Indomie'],
    'style': ['cup', 'cup', 'cup', 'pack', 'pack'],
    'rating': [4, 4, 3.5, 15, 5]
})
df
```

```python
df.drop_duplicates() #entire same will droped
```

```python
df.drop_duplicates(subset=['brand', 'style'], keep='last')# remove duplicates on specific column(s), use subset.
#To remove duplicates and keep last occurrences, use keep.
```
